```python
import jieba
import os
texts = os.listdir('C:/Users/user/Desktop/HW0116-1/')
texts
```

```python
open('./HW0116-1/'+texts[0], encoding='utf-8').read()
```

```python
corpus = []
for text in texts:
    with open('C:/Users/user/Desktop/HW0116-1/{}'.format(text), 'r', encoding='utf-8') as file:
        corpus.append(file.read())
```

```python
corpus_cut = [jieba.lcut(t) for t in corpus]
# 這樣就有斷完詞後的 list 了
corpus_cut
```

```python
# 斷詞後的 list 變成用空格把每個詞分開的形式
corpus = [' '.join(lcut) for lcut in corpus_cut]
# 然後就可以接到 apriori 的過程了
corpus
```
